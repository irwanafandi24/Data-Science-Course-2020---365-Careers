# Simple Linear Regression NN

## Import the relevant libraris

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D
```

## Generate random input data to train on 

```python
observations = 1000
xs = np.random.uniform(low=-10, high=10, size=(observations, 1)) #shape 1000 and 1 rows
zs = np.random.uniform(low=-10, high=10, size=(observations, 1))

inputs = np.column_stack((xs,zs)) #marge 2 data into songle data, shape 1000 and 2 rows
inputs.shape
```

## Create the target 
f(x,z) = 2*x - 3*z + 5 + noise<br>
w1 = x<br>
w2 = z<br>
bias = 5

```python
noise = np.random.uniform(-1, 1, (observations,1))
target = 2*xs-3*zs+5+noise
target.shape
```

## Plot the training data 

```python
target = target.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs,zs, target)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('target')
ax.view_init(azim=100) #plot the data from different angle
plt.show()
target = target.reshape(observations,1)
```

## Initial variable 

```python
init_range = 0.1 #start from small one
weights = np.random.uniform(-init_range, init_range, size=(2,1)) # we have 2 variable
bias = np.random.uniform(-init_range, init_range, size=1) #scalar

print(weights, bias)
```

## Set the learning rate (eta) 

```python
learning_rate = 0.02
```

## Train the model 
Create the loop which will apply update rule and calculate the loss function<br>
for each loop
1. Calculate the output
2. Compare output with the target through the loss
3. Print the loss, L2-norm loss formula: sigmai (yi-ti)**2
4. Adjust(update) weights and biases, formula : wi+1 = wi - eta x sum of xi * deltai || bi+1 = bi - eta x sum of deltai

```python
for i in range(10000):
    #Output (1000, 1) = input (1000,2) x weights (2,1) + bias (scalar)
    output = np.dot(inputs, weights) + bias #y = wx+b
    #deltas (1000,1)
    deltas = output - target
    #get the loss L2-norm loss, devided by 2 to get the elegant update rules from the gradient descent
    #devided by observations to get the average loss per observation
    loss = np.sum(deltas ** 2) / 2 / observations
    print(loss)
    
    #we need scaled the delta to make the algorithm more universal
    delta_scaled = deltas / observations
    #using gradient descend metodology
    #inputs (1000,2), deltas(1000,1) dot product will never works. So we need to transpose the inputs
    weights = weights - learning_rate * np.dot(inputs.T, delta_scaled) #note: dot product == sum of product
    bias = bias - learning_rate * np.sum(delta_scaled)
```

```python
#print the weights and bias
print(weights,bias)
```
