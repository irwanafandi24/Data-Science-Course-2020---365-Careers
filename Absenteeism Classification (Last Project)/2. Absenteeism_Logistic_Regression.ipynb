## Standardized the data 
- In this case, we need to scale unscaled_inputs, but without include the reason feature. Use the function Customer scaller 

```python
#this code is used for scaled all the data
# from sklearn.preprocessing import StandardScaler
# scale = StandardScaler()
# scale.fit(unscaled_inputs)

# scaled_inputs = scale.transform(unscaled_inputs)
# scaled_inputs
```

```python
#with the code above, we will standarize all the data include dummy. But we don't want because the main
#reason information gonna be loss, so we need to make custome standardization
from sklearn.preprocessing import StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin

class CustomerScaler(BaseEstimator, TransformerMixin):
    def __init__(self, columns, copy=True, with_mean=True, with_std= True):
        self.scaler = StandardScaler(copy, with_mean, with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
#our CustomerScaler function needs the column's name that want to scaled, so we need to declare it first
unscaled_inputs.columns
```

```python
#choose the column (before backward elimination)
# column_to_scale = ['Month','Day of the week', 'Transportation Expense', 'Distance to Work', 'Age',
#        'Daily Work Load Average', 'Body Mass Index', 'Children','Pets']

#after backward elimination implemented
column_to_scale = ['Day of the week', 'Transportation Expense', 'Age',
       'Body Mass Index', 'Children','Pets']
#start to scale the data
absenteeism_scaler = CustomerScaler(column_to_scale)
```

```python
#see the results, the reasons and educations columns do not scale
absenteeism_scaler.fit(unscaled_inputs)

scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
scaled_inputs.head()
```

## Split the data 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_inputs, targets, test_size=0.2, random_state= 27)
```

## Build Model Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

## Get the intercept and coefficient 

```python
model.intercept_
```

```python
model.coef_
```

```python
unscaled_inputs.columns
```

```python
summary_table = pd.DataFrame(columns=['Feature Name'], data = unscaled_inputs.columns)
summary_table['Coef'] = np.transpose(model.coef_)
summary_table.sort_values(by=['Coef'], ascending=False)
```

```python
#add intercept values to the list
summary_table.index = summary_table.index+1
summary_table.loc[0] = ['Intercept', model.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
#when we dealing with logistic regression, the coefficient we are predicting called log(odds), 
#in logistic regression, the log(odds) transform to 0 and 1
#log(odds) = bias(intercept) + coef1.x1+ coef2.x2 +  ... + coefn.xn
#to make them more interpretable, let's find the exponentials of these coefficients

summary_table['Odds_ratio'] = np.exp(summary_table['Coef'])
summary_table
```

```python
#The further away from 0 a coefficient is, the bigger its importance  
summary_table.sort_values(by=['Odds_ratio'], ascending = False)
```
